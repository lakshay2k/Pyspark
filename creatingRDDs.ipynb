```
#Creating RDDs using parallelize function
```

```
from pyspark.sql import SparkSession
spark = SparkSession \
.builder \
.appName("Python Spark create RDD example") \
.config("spark.some.config.option", "some-value") \
.getOrCreate()
df = spark.sparkContext.parallelize([(1, 2, 3, 'a b c'),
(4, 5, 6, 'd e f'),
(7, 8, 9, 'g h i')]).toDF(['col1', 'col2', 'col3','col4'])
```

```
df.show()
```

```
mylist = [['Male',1,None],['Female',2,3],['Male',3,4]]
ds = spark.createDataFrame(mylist,['col1','col2','col3'])
ds.show()
```

```
ds.fillna(-99).show()
```

```
ds.na.replace(['Male','Female'],['1','0']).show()
```

```
ds.toDF("C1","C2","C3").show()
```

```
#Renaming the columns
```

```
mapping = {"col2":"Column2","col3":"Column3"}
new_names = [mapping.get(col,col) for col in ds.columns]
ds.toDF(*new_names).show()
```

```
ds.withColumnRenamed('col1','Column1').show()
```

```
#dropping columns
```

```
drop_names = ['col2','col3']
ds.drop(*drop_names).show()
```
