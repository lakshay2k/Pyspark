```
from pyspark.sql import SparkSession

spark = SparkSession \
        .builder \
        .appName("RDD Operations") \
        .config("spark.some.config.option","some-value") \
        .getOrCreate()

df = spark.read.format('com.databricks.spark.csv').\
options(header='true', \
inferschema='true').\
load("dbfs:/FileStore/tables/movies.txt",header=True)
df.show(5)
df.printSchema()
```

```
import pandas as pd
```

```
leftp = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['B0', 'B1', 'B2', 'B3'],
'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']},
index=[0, 1, 2, 3])
rightp = pd.DataFrame({'A': ['A0', 'A1', 'A6', 'A7'],
'F': ['B4', 'B5', 'B6', 'B7'],
'G': ['C4', 'C5', 'C6', 'C7'],
'H': ['D4', 'D5', 'D6', 'D7']},
index=[4, 5, 6, 7])

lefts = spark.createDataFrame(leftp)
rights = spark.createDataFrame(rightp)

lefts.show()
```

```
rights.show()
```

```
#performing Left join on two rdds
```

```
rdd1=lefts.join(rights, on='A', how='left') \
     .orderBy('A',ascending=True)
rdd1.show()
```

```
#Right join
```

```
rdd2 = lefts.join(rights,on='A',how='right') \
            .orderBy('A',ascending=True)
rdd2.show()
```

```
#Inner join
```

```
rdd3 = lefts.join(rights,on='A',how='inner') \
            .orderBy('A',ascending=True)
rdd3.show()
```

```
#Full Join
```

```
rdd4 = lefts.join(rights,on='A',how='full') \
            .orderBy('A',ascending=True)
rdd4.show()
```
